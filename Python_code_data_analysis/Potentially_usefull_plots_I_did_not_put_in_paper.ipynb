```python
import numpy as np
import csv
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
mu = '0.05'
gamma = '3'
path = r'C:\Users\Wieke Joustra\Documents\Bachelor\BEP\cell_sorting_with_cc3d\Final_Results\Mu_' + mu + '\gamma_' + gamma

# filearray = [path + '\\numberPcells_run1.csv', path + '\\numberPcells_run5.csv', path + '\\numberPcells_run9.csv', 
#             path + '\\numberPcells_run13.csv', path + '\\numberPcells_run17.csv']

filearray = [path + '\\numberPcells_run2.csv', path + '\\numberPcells_run6.csv', path + '\\numberPcells_run10.csv', 
            path + '\\numberPcells_run14.csv']


finalMCS = 200000 
output_frequency = 1000
xaxis = int(finalMCS/output_frequency)


a=1
for j in filearray: 
    MCS = []
    cellid = []
    xCOM = []
    yCOM = []
    zCOM = []
    with open(j) as f: 
        reader = csv.reader(f, delimiter= ' ')
        for column in reader:
            if column != 0:
                for row in reader:
                    MCS.append(float(row[0]))
                    cellid.append(float(row[1]))
                    xCOM.append(float(row[2]))
                    yCOM.append(float(row[3]))
                    zCOM.append(float(row[4]))
                    
    print('ik ga nu de loop in')

    for i in range(0, xaxis): #loopen over de hoeveelheid cellen (kan ook MCE_P gebruiken: deze zijn hetzelfde)
        if i % 10 ==0: 
            xCOM_per_cell = [] #elke cell een lege array
            yCOM_per_cell = []
            zCOM_per_cell = []
            for k in range(0, len(cellid)): #dan over alle *waardes* van aaaalle cell-id's loopen
                if k % 10 == 0: 
                    if cellid[k] == cellid[i]:  #en als deze k-loop-cell gelijk is aan de oorspronkelijke i-loop cell
                                                #dus als de k-cell-id gelijk is aan de i cell id
                        xCOM_per_cell.append(xCOM[k]) #dan xco, yco, zco vaststellen
                        yCOM_per_cell.append(yCOM[k])
                        zCOM_per_cell.append(zCOM[k])

            fig = plt.figure(a, figsize=(20,40))
            titel = fig.suptitle('Run numer is: ' + str(a))
            plt.subplot(531)       
            plt.plot(xCOM_per_cell,yCOM_per_cell)
            plt.xlabel('xCOM'), plt.ylabel('yCOM')
            plt.subplot(532)
            plt.plot(xCOM_per_cell,zCOM_per_cell)
            plt.xlabel('xCOM'), plt.ylabel('zCOM')
            plt.subplot(533)
            plt.plot(yCOM_per_cell,zCOM_per_cell)
            plt.xlabel('yCOM'), plt.ylabel('zCOM')

    a+=1
```
